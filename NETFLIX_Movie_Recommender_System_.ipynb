Feature Selection

```python
df=df[['id','title','overview','genre']]
df.head()
```

```python
df['tags'] = df['overview'] + df['genre']
df.head()
```

```python
new_data=df.drop(columns=['overview','genre'])
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv =CountVectorizer(max_features=1000,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

Cosine_Similarity


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python

similarity
```

Access the id and index number

```python
new_data[new_data['title']=="The Godfather"]
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda vector:vector[1])
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python

recommand("Iron Man")
```
